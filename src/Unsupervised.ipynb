# Wine Dataset
##K-Means
### Perform Kmeans analysis
We will start by analyzing different cluster values for Kmeans and see how well they perform when compared to the following scoring tests. The ideas here is to vary our cluster size and then use a few metrics to gauge what our ideal k cluster size should be set to.
- Homogeneity
- Completeness
- V-measure
- Silhouette Score


```python
hArr = []
cArr = []
vArr = []
sArr = []
numPoints = 40
for i in range(2, numPoints):
    km = KMeans(n_clusters=i, max_iter=300)
    km.fit(X_train)
    hArr.append(metrics.homogeneity_score(y_train, km.labels_))
    cArr.append(metrics.completeness_score(y_train, km.labels_))
    vArr.append(metrics.v_measure_score(y_train, km.labels_))
    sArr.append(metrics.silhouette_score(X_train, km.labels_, metric='euclidean'))
    
x = xrange(2, numPoints)
fig = plt.figure()
plt.plot(x, hArr, label='Homogeneity Score')
plt.plot(x, cArr, label='Completeness Score')
plt.plot(x, vArr, label='V-measure Score')
plt.plot(x, sArr, label='Silhouette Score')
plt.legend(loc='upper center', shadow=True)
fig.show()

```

##Analysis of K-Means cluster sizing
After analyzing the different cluster sizes for the k-means algorithm it was determined that the best overall combination for our scores seemed to happen with a k of around 11. This will be the cluster size that will be used for the next experiment. This makes some empirical sense as this is close to the range of values that the wine experts applied as scores to the individual wines in the dataset

## Expectation Maximization
### Initial Work
For the inital part of the Expectation Maximization analysis I chose to follow a similar approach taken in the K-means exaple. I have chosed to create a Scatterplot matrix. Each of the child plots in the matrix will give a visualization of liklehoods as predicted by an Expectation Maximization algorithm [Gaussian Mixture Models](http://scikit-learn.org/stable/modules/mixture.html) in this case

In this section I attempted to get a few contour plots with a few 2d combinations of attributes. This was done to get a feeling for which attribute combinations were providing the best clustering

```python
#Normalize X_train
X_normalized = preprocessing.scale(X_train)

a = X_normalized[:,winedata['pH']]
b = X_normalized[:,winedata['alcohol']]
c = X_normalized[:,winedata['sulphates']]
d = X_normalized[:,winedata['fixed acidity']]
X_subset = []
X_subset.append(np.column_stack((a,b)))
X_subset.append(np.column_stack((c,b)))
X_subset.append(np.column_stack((c,a)))
X_subset.append(np.column_stack((a,d)))

for x_entry in X_subset:
    # fit a Gaussian Mixture Model with two components
    clf = mixture.GMM(n_components=10, covariance_type='diag')
    clf.fit(x_entry)

    # display predicted scores by the model as a contour plot
    x = np.linspace(x_entry[:, 0].min(), x_entry[:, 0].max())
    y = np.linspace(x_entry[:, 1].min(), x_entry[:, 1].max())
    X, Y = np.meshgrid(x, y)
    XX = np.array([X.ravel(), Y.ravel()]).T
    Z = -clf.score_samples(XX)[0]
    Z = Z.reshape(X.shape)

    CS = plt.contour(X, Y, Z, norm=LogNorm(vmin=1.0, vmax=1000.0),
            levels=np.logspace(0, 3, 10))
    CB = plt.colorbar(CS, shrink=0.8, extend='both')
    plt.scatter(x_entry[:, 0], x_entry[:, 1], .8)

    plt.title('Negative log-likelihood predicted by a GMM')
    plt.axis('tight')
    plt.figure(figsize=(20,20))
    plt.show()


```

### Initial K-means
For this part of the assignment the sepal length, petal width, and petal length were chosen as attributes to create our initial k-means visualization from. 

In the graphs and scoring table below we can see that although using 8 clusters for k-means produces tighter clusters it appears that setting the cluster number to 3 produces better results. If we take a look at the silhouette score we can see that the value goes up dramatically for the 3 cluster case. Intuitivley this is what we expect as we have 3 classes of flower that are used in the dataset and we would imagine that the different flowers have different physical properties when it comes to length and width that would allow them to be categorized quite well.

```python
centers = [[1, 1], [-1, -1], [1, -1]]
iris = datasets.load_iris()
X = iris.data
y = iris.target

estimators = {'k_means_iris_3': KMeans(n_clusters=3),
              'k_means_iris_8': KMeans(n_clusters=8)}

fignum = 1
for name, est in estimators.items():
    fig = plt.figure(fignum, figsize=(8, 6))
    plt.clf()
    ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)

    plt.cla()
    est.fit(X)
    labels = est.labels_
    
    print("Homogeneity: %0.3f" % metrics.homogeneity_score(y, labels))
    print("Completeness: %0.3f" % metrics.completeness_score(y, labels))
    print("V-measure: %0.3f" % metrics.v_measure_score(y, labels))
    print("Silhouette Score: %0.3f\n" % metrics.silhouette_score(X, labels, metric='euclidean'))

    ax.scatter(X[:, 3], X[:, 0], X[:, 2], c=labels.astype(np.float))

    ax.w_xaxis.set_ticklabels([])
    ax.w_yaxis.set_ticklabels([])
    ax.w_zaxis.set_ticklabels([])
    ax.set_xlabel('Petal width')
    ax.set_ylabel('Sepal length')
    ax.set_zlabel('Petal length')
    fignum = fignum + 1

# Plot the ground truth
fig = plt.figure(fignum, figsize=(8, 6))
plt.clf()
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)

plt.cla()

for name, label in [('Setosa', 0),
                    ('Versicolour', 1),
                    ('Virginica', 2)]:
    ax.text3D(X[y == label, 3].mean(),
              X[y == label, 0].mean() + 1.5,
              X[y == label, 2].mean(), name,
              horizontalalignment='center',
              bbox=dict(alpha=.5, edgecolor='w', facecolor='w'))
# Reorder the labels to have colors matching the cluster results
y = np.choose(y, [1, 2, 0]).astype(np.float)
ax.scatter(X[:, 3], X[:, 0], X[:, 2], c=y)

ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])
ax.set_xlabel('Petal width')
ax.set_ylabel('Sepal length')
ax.set_zlabel('Petal length')
plt.show()
```

### Expectation Maximization

```python
iris = datasets.load_iris()

# Break up the dataset into non-overlapping training (75%) and testing
# (25%) sets.
skf = StratifiedKFold(iris.target, n_folds=4)
# Only take the first fold.
train_index, test_index = next(iter(skf))


X_train = iris.data[train_index]
y_train = iris.target[train_index]
X_test = iris.data[test_index]
y_test = iris.target[test_index]

n_classes = len(np.unique(y_train))

# Try GMMs using different types of covariances.
classifiers = dict((covar_type, mixture.GMM(n_components=n_classes,
                    covariance_type=covar_type, init_params='wc', n_iter=20))
                   for covar_type in ['spherical', 'diag', 'tied', 'full'])

n_classifiers = len(classifiers)

plt.figure(figsize=(9 * n_classifiers / 2, 18))
plt.subplots_adjust(bottom=.01, top=0.95, hspace=.15, wspace=.05,
                    left=.01, right=.99)


for index, (name, classifier) in enumerate(classifiers.items()):
    # Since we have class labels for the training data, we can
    # initialize the GMM parameters in a supervised manner.
    classifier.means_ = np.array([X_train[y_train == i].mean(axis=0)
                                  for i in xrange(n_classes)])

    # Train the other parameters using the EM algorithm.
    classifier.fit(X_train)

    h = plt.subplot(2, n_classifiers / 2, index + 1)
    plotFunctions.make_ellipses(classifier, h)

    for n, color in enumerate('rgb'):
        data = iris.data[iris.target == n]
        plt.scatter(data[:, 0], data[:, 1], 0.8, color=color,
                    label=iris.target_names[n])
    # Plot the test data with crosses
    for n, color in enumerate('rgb'):
        data = X_test[y_test == n]
        plt.plot(data[:, 0], data[:, 1], 'x', color=color)

    y_train_pred = classifier.predict(X_train)
    train_accuracy = np.mean(y_train_pred.ravel() == y_train.ravel()) * 100
    plt.text(0.05, 0.9, 'Train accuracy: %.1f' % train_accuracy,
             transform=h.transAxes)

    y_test_pred = classifier.predict(X_test)
    test_accuracy = np.mean(y_test_pred.ravel() == y_test.ravel()) * 100
    plt.text(0.05, 0.8, 'Test accuracy: %.1f' % test_accuracy,
             transform=h.transAxes)

    plt.xticks(())
    plt.yticks(())
    plt.title(name)

plt.legend(loc='lower right', prop=dict(size=12))


plt.show()
```
