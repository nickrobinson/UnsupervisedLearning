### Perform Kmeans analysis
We will start by analyzing different cluster values for Kmeans and see how well they perform when compared to the following scoring tests.
- Homogeneity
- Completeness
- V-measure
- Silhouette Score


```python
hArr = []
cArr = []
vArr = []
sArr = []
for i in range(1,40):
    km = KMeans(n_clusters=i, max_iter=300)
    km.fit(X_train)
    hArr.append(metrics.homogeneity_score(y_train, km.labels_))
    cArr.append(metrics.completeness_score(y_train, km.labels_))
    vArr.append(metrics.v_measure_score(y_train, km.labels_))
    sArr.append(metrics.silhouette_score(X_train, km.labels_, metric='euclidean'))
    
x = xrange(39)
fig = plt.figure()
plt.plot(x, hArr)
plt.plot(x, cArr)
plt.plot(x, vArr)
fig.show()

```
